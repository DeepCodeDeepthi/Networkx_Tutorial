# Labelling nodes in a graph = City Network

```python
import networkx as nx
```

```python
city = [ 'Delhi','Mumbai' , 'Lucknow' , 'Vizag' , 'Bangalore' ]
```

```python
city
```

```python
G=nx.complete_graph(city)
```

```python
G
```

```python
nx.draw(G,with_labels=True , node_color='y')
```

```python
G.nodes()
```

```python
G.edges()
```
