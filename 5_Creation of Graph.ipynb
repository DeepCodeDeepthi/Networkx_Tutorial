# Creation OF Graph :

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G
```

### Node Creation :

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
```

### Edge Creation :

```python
G.add_edge(1,2)
G.add_edge(1,3)
G.add_edge(2,3)
G.add_edge(2,4)
G.add_edge(3,4)
G.add_edge(4,1)
```

### Draw Graph :

```python
nx.draw(G , with_labels = True , node_color='yellow' , font_weight='bold')
```
